```python
import pandas as pd
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
# Histograma de Pandas
df['math score'].plot.hist()   # Histograma
```

```python
# Histograma de Seaborn
import seaborn as sns
sns.distplot(df['math score'])     # Contiene curvas de Kernel
```

```python
# Histograma alternativo a Seaborn
sns.displot(df['math score'])        # Sin la 'T'
```

```python
sns.displot(df['math score'], kde = True, bins = 10)      # De esta forma le activamos la curva de Kernel y le damos el numero de bins
```

```python
sns.distplot(df['math score'], hist = False) # Con DISTTTTPLOT le podemos quitar el histograma para solo ver la curva de kernel
```

```python
sns.distplot(df['math score'], hist = False, kde_kws={'shade' : True})  # De esta forma colorea la curva de kernel
```

```python
# Creacion de diagramas de dispercion con recta de regrecion  (comparacion de dos puntos entre si)
# Tendencia de relacion de dos puntos entre si
sns.regplot(x='reading score', y= 'writing score', data = df)
```

```python
sns.lmplot(x='reading score', y= 'writing score', data = df)
# Esta es una mejor grafica de dispercion
```

```python
# Grafico de dispercion con regrecion lineal por categoridad de genero

sns.lmplot(x='reading score', y= 'writing score', data = df, hue = 'gender')
```

```python
sns.lmplot(x='reading score', y= 'writing score', data = df, col = 'gender')
```

```python
sns.lmplot(x='reading score', y= 'writing score', data = df, col = 'race/ethnicity')
```
