```python
import pandas as pd
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
import seaborn as sns
for style in ['dark', 'whitegrid', 'darkgrid', 'ticks']:
    sns.set_style(style)
    sns.displot(df['reading score'])
# 1ero 'DARK'. 2do 'WHITEGRID'. 3ro 'DARKGRID'. 4to 'TICKS'.
```

```python
# Remocion de ejes
sns.set_style('ticks')
sns.displot(df['reading score'], kde = True)
sns.despine(left=True)      # Se eleimina el eje vertical
```

```python
sns.set_style('ticks')
sns.displot(df['reading score'], kde = True)
sns.despine(left=True, bottom=True)  # Asi se eliminan los dos ejes tanto el vertical y el horizontal
```

```python
# Combinacion de parametros con Matplotib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sns.distplot(df['writing score'], ax = ax)
ax.set(xlabel = 'Score de escritura', ylabel = 'frecuencia relativa')
```

```python
# Trncamiento de escala de visualizacion
# poner los limites del EJE X
# Y colocacion de un titulo
fig, ax = plt.subplots()
sns.distplot(df['writing score'], ax = ax)
ax.set(xlabel = 'Score de escritura', ylabel = 'frecuencia relativa', xlim = (0,90),
       title = 'Histograma de frecuencias para puntajes de escritura')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows= 1, ncols= 2, sharey= True)
sns.distplot(df['writing score'], ax = ax0)  # Histograma para todas las observaciones
sns.distplot(df.query('gender == "male"')['writing score'], ax = ax1) # Histograma para los hombres
ax0.set(xlabel = 'Puntuaje generL de escritura')
ax1.set(xlabel = 'Puntuaje de escritura para los hombres')
ax1.axvline(x = 75, label = 'objetivo', linestyle = '--')
ax1.legend()
```

```python
# Asignacion de colores al histograma
paleta = ['deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind']

for p in paleta:
    sns.set_palette(p)
    sns.displot(df['reading score'], kde = True)
```
